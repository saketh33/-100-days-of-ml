```python
import pandas as pd
df = pd.read_csv('C:\\Users\\saketh\\Documents\\boston.csv')
print(df.shape)
df.head()
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
dft = AV.AutoViz(filename='', sep=',', depVar='medv', dfte=df, header=0, verbose=2,
                            lowess=False,chart_format='svg',max_rows_analyzed=1500,max_cols_analyzed=30)
```
